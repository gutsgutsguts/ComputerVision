numpy 배열 정보

ndarray.ndim : 배열 차원의 축 갯수

ndarray.shape : 배열의 각차원을 튜플로 나타낸 것

ndarray.dtype : 배열 요소 자료형

ndarray.itemsize : 각 요소의 바이트 사이즈



```python
print("vector]","\n","-ndim:",vector.ndim,"-shape:",vector.shape,"-size:",vector.size,"-dtype:",vector.dtype,"-itemsize:",vector.itemsize)
print(vector)
print("\n"+"matrix]","\n","-ndim:",matrix.ndim,"-shape:",matrix.shape,"-size:",matrix.size,"-dtype:",matrix.dtype,"-itemsize:",matrix.itemsize)
print(matrix)
print("\n"+"tensor]","\n","-ndim:",tensor.ndim,"-shape:",tensor.shape,"-size:",tensor.size,"-dtype:",tensor.dtype,"-itemsize:",tensor.itemsize)
print(tensor)
```

Numpy 인덱싱, 슬라이싱

슬라이싱

   &nbsp; 배열 [a:b]를 이용하여 배열을 자를 수 있음
  
  &nbsp; a: 시작 , b : 도착(슬라이싱한 값에 포함 되지 않음)

index

&nbsp;  컴퓨터는 항상 0 부터 시작
  
&nbsp;  그러므로 끝은 총 갯수의 -1

&nbsp;  숫자 없이 : 사용경우, 모든 행 또는 열을 의미

&nbsp; [:n] 은 0 부터 n-1 까지

&nbsp; [n:] d은 n 부터 끝까지



```python
data = np.array(["a", "b", "c"])

print(" data\n",data)
print("\n","data[0]\n",data[0])
print("\n","data[1]\n",data[1])
print("\n","data[0:2]\n",data[0:2])
print("\n","data[1:]\n",data[1:])
print("\n","data[-2:]\n",data[-2:])

```

numpy 배열간 브로드캐스팅 사칙연산

작은 크기의 배열을 숫자가 큰 크기의 배열에 흩뿌려지는연산

조건 1 : 각 차원 의 크기가 같을때

조건 2 : 배열 크기 다르더라도 각 차원의 크기가 둘중 하나 1인 경우



```python
data = np.array([3, 4])

ones = np.ones(2, dtype=int)

print("data\n",data)
print("\n","ones\n",ones)
print("\n","data + ones\n",data + ones)
print("\n","data - ones\n",data - ones)
print("\n","data * ones\n",data * ones)
print("\n","data / ones\n",data / ones)
```

```python
data = np.array([[1,2], [3, 4]])
data2 = np.array([[12,13] , [14,16]])

print("\n","data + data2\n",data + data2)
```

```python
data1D = np.array([1.0, 2.0, 3.0])
print("data1D*1.6" , "\n", data1D*1.6, "\n")

data2D = np.array([[1,2], [3,4], [5,6]])
ones_row = np.array([[1,1]])
print("data2D+ones_row" , "\n", data2D+ones_row, "\n")
```

```python
import numpy as np
import matplotlib.pyplot as plt

#300행, 300열, 3개의 RGB 채널, 픽셀 당 8 bits 크기를 가진 검은색과 하얀색 컬러 이미지를 생성
canvas_b = np.zeros((300,300,3), np.uint8)
canvas_w = np.full((300,300,3), 255, np.uint8)

images = [canvas_b, canvas_w]
titles = ['Black Image','White Image']

fig, axes = plt.subplots(2, 1) # 2개의 그래프가 들어가는 Figure 생성
for i in range(len(images)):
    plt.subplot(1,2,i+1)
    plt.imshow(images[i])
    plt.title(titles[i])

plt.savefig("figure2", transparent=True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 캔버스 생성
canvas_b = np.zeros((300, 300, 3), np.uint8)
canvas_w = np.full((300, 300, 3), 255, np.uint8)

images = [canvas_b, canvas_w]
titles = ['black', 'white']

# 2행 1열의 서브플롯 생성
fig, axes = plt.subplots(1, 2, figsize=(8, 4))

# 각 서브플롯에 이미지와 제목 설정
for i in range(len(images)):
    axes[i].imshow(images[i])
    axes[i].set_title(titles[i])
    #axes[i].axis('off')  # 축 표시를 끔

# 이미지 저장 및 출력
plt.tight_layout()  # 레이아웃 자동 조정
plt.savefig("canvas.png", transparent=True)
plt.show()
```

open cv 로 그림 그리기

선

image - 그림을 그릴 이미지 파일

start - 시작점 좌표

end - 끝 좌표

color - RGB 형태의 컬러

thickness - 두께

linetype - 선의 종류

```python
import cv2
from google.colab.patches import cv2_imshow #colab 호환 가능한 imshow 함수
import numpy as np
import matplotlib.pyplot as plt

canvas_b = np.zeros((300, 300, 3), np.uint8)
canvas_w = np.full((300, 300, 3), 255, np.uint8)

opencv_red = (0, 0, 255)
opencv_green = (0, 255, 0)
opencv_blue = (255, 0, 0)

cv2.line(canvas_b, (50, 50), (300, 300), opencv_red, 5)
cv2.line(canvas_b, (50, 100), (250, 100), opencv_green, 15)

cv2.line(canvas_w, (0, 0), (250, 150), opencv_red, 32, cv2.LINE_AA)
cv2.line(canvas_w, (50, 200), (250, 200), opencv_blue, 16, cv2.LINE_4)
cv2.line(canvas_w, (0, 250), (250, 250), opencv_green, 16, cv2.LINE_8)

cv2_imshow(canvas_b)
cv2_imshow(canvas_w)
```

```python
images = [canvas_b, canvas_w]
titles = ['Black Image','White Image']

for i in range(len(images)):
    plt.subplot(1,2,i+1),plt.imshow(images[i])
    plt.title(titles[i])

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

canvas_w = np.full((300,300,3), 255, np.uint8)
cv2.rectangle(canvas_w,(0,0),(299,299),opencv_red)
cv2.rectangle(canvas_w,(10,10),(80,80),opencv_green,4)
cv2.rectangle(canvas_w,(80,80),(180,150),opencv_blue,-1)

cv2.circle(canvas_w,(180,150),70,opencv_red)
cv2.ellipse(canvas_w,(180,150),(50,100),0,0,360,opencv_green,4)
cv2.ellipse(canvas_w,(210,210),(50,100),45,180,360,opencv_blue,4,cv2.LINE_AA)


triangle1 = np.array([[10,200],[10, 240],[100, 220]])
triangle2 = np.array([[10,260],[10, 290],[100, 290]])
cv2.polylines(canvas_w, [triangle1], True, opencv_red,1)
cv2.polylines(canvas_w, [triangle2], False, opencv_red,4)


#opencv 색상을 opencv에서 표현
cv2_imshow(canvas_w)
#opencv 색상을 matplotlib에서 표현
plt.imshow(canvas_w)
plt.show()
```
